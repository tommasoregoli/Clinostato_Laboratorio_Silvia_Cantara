```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Raw_file = pd.read_excel("./Clinostate_Accelerometer_results_18_6_26.xlsx")

x = Raw_file['AccX(g)'] 
y = Raw_file['AccY(g)']
z = Raw_file['AccZ(g)']
```

```python
# Create the figure Area and axis
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# Print Points (Scatter Plot)
# s=2 dot size, alpha=0.4 is trasparency value
ax.scatter(x, y, z, c='green', marker='o', s=2, alpha=0.4)

# Insert title and labeling 
ax.set_title('3D Acceleration Distribution - 1 hour', fontsize=14)
ax.set_xlabel('X Axis (g)')
ax.set_ylabel('Y Axis (g)')
ax.set_zlabel('Z Axis (g)')

# Insert axis limits
ax.set_xlim([-1.0, 1.0])
ax.set_ylim([-1.0, 1.0])
ax.set_zlim([-1.0, 1.0])

# Show results
plt.tight_layout()
plt.subplots_adjust(right=2.85)
plt.show()
```
